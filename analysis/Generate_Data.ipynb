```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = []

for i in range(10000):

    hour = i % 24

    # Day/night light cycle
    light = 50 + 35*np.sin(2*np.pi*hour/24) + np.random.normal(0,10)

    # Natural environmental variation
    soil = np.random.normal(50,15)
    humidity = np.random.normal(45,12)

    # Limits
    soil = np.clip(soil,10,90)
    humidity = np.clip(humidity,20,80)
    light = np.clip(light,0,100)

    # Stress score
    stress = (
        0.45*(1-soil/100) +
        0.30*(1-humidity/100) +
        0.25*(1-light/100)
    )

    stress += np.random.normal(0,0.06)
    stress = np.clip(stress,0,1)

    # Health score
    health = (
        0.55*soil +
        0.20*humidity +
        0.15*light +
        10*(1-stress)
    )

    health += np.random.normal(0,4)
    health = np.clip(health,0,100)

    # Probabilistic alert label
    risk = (
        0.5*(1-soil/100) +
        0.25*(1-humidity/100) +
        0.25*stress
    )

    risk += np.random.normal(0,0.08)

    prob_alert = 1 / (1 + np.exp(-9*(risk - 0.50)))

    alert = int(np.random.rand() < prob_alert)

    rows.append([soil,humidity,light,stress,health,alert])

df = pd.DataFrame(rows, columns=[
    "soil","humidity","light","stress","health","alert"
])

df.to_csv("ml_training_data.csv", index=False)

print(df.head())
print("Rows:", len(df))
```
